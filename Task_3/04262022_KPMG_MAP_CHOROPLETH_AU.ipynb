```python
#libraries

import numpy as np
import pandas as pd
import geopandas as gpd
import json
from urllib.request import urlopen
import plotly.express as px
import plotly.graph_objects as go
```

## Choropleth Map

```python
#get AU geometry
with urlopen('https://raw.githubusercontent.com/rowanhogan/australian-states/master/states.geojson') as response:
    counties = json.load(response)


    
#counties["features"][0]['properties'].keys()
#counties["features"][0]
```

```python
counties["features"][0]['properties'].keys()
```

```python
#getting stat data

stat = pd.read_csv("./data/stats_bike_17.csv")
stat = stat.rename(columns={'name':'STATE_NAME'})
```

```python
stat
```

```python
#Plotting data // CHANGE THE VAR BY A BOX_SELECTION

max_color=stat.total.max()
min_color=stat.total.min()

fig = go.Figure(go.Choroplethmapbox(geojson=counties, 
                                    locations=stat.STATE_NAME, 
                                    z=stat.total,
                                    text=stat.state,
                                    featureidkey="properties.STATE_NAME",
                                    colorscale="Redor", 
                                    zmin=min_color, 
                                    zmax=max_color,
                                    marker_opacity=0.8, 
                                    marker_line_width=0,
                                    name="Total Sample (n)",
                                    hovertemplate="%{text}<br>%{z}%"))

fig.update_layout(mapbox_style="carto-positron",
                  mapbox_zoom=2.6, 
                  mapbox_center = {"lat": -25.2744, "lon": 133.7751},
                 mapbox_layers=[fill="#fff"])
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.update_traces(colorbar=dict(len=0.5,
                                ticks='inside',
                                thickness=10,
                                ticklen=0,
                                outlinewidth=0),
                  selector=dict(type='choroplethmapbox'))
```
